## What proportion of each individual charge made up the total bill 
. . . and how did those proportions change over time?
1. `service_charge`
2. `consumption_charge`
3. `peak_charge`
4. `fuel_charge`
5. `regulatory_charge`

```python
df.head(2)
```

```python
df['service_charge_%'] = df.service_charge / df.total_bill
df['consumption_charge_%'] = df.consumption_charge / df.total_bill
df['peak_charge_%'] = df.peak_charge / df.total_bill
df['fuel_charge_%'] = df.fuel_charge / df.total_bill
df['regulatory_charge_%'] = df.regulatory_charge / df.total_bill

df
```

```python
df.set_index('bill_date')['consumption_charge_%'].plot()
plt.title('% Consumption of Total Bill Over Time')
plt.show()
```

```python
df.set_index('bill_date')['fuel_charge_%'].plot()
plt.title('% Fuel Surcharge of Total Bill Over Time')
plt.show()
```

```python
for bill_part_pct in df.iloc[:, -5:].columns:
    df.set_index('bill_date')[bill_part_pct].plot()
    plt.title(f'{bill_part_pct} of Total Bill Over Time')
    plt.show()
    print('************************************************************************')
```
