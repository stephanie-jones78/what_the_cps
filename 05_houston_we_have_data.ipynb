```python
import pandas as pd

# show all columns
pd.set_option('display.max_columns', None)

import wrangle

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# storing august and september bill filepaths (september is for testing)
filePath_sep21 = 'pdfs/2021-09-28_bill.pdf'
filePath_oct21 = 'pdfs/2021-10-27_bill.pdf'
filePath_nov21 = 'pdfs/2021-11-29_bill.pdf'
filePath_dec21 = 'pdfs/2021-12-28_bill.pdf'
filePath_jan22 = 'pdfs/2022-01-28_bill.pdf'
filePath_feb22 = 'pdfs/2022-02-25_bill.pdf'
filePath_mar22 = 'pdfs/2022-03-25_bill.pdf'
filePath_apr22 = 'pdfs/2022-04-27_bill.pdf'
filePath_may22 = 'pdfs/2022-05-26_bill.pdf'
filePath_jun22 = 'pdfs/2022-06-27_bill.pdf'
filePath_jul22 = 'pdfs/2022-07-27_bill.pdf'
filePath_aug22 = 'pdfs/2022-08-25_bill.pdf'
filePath_sep22 = 'pdfs/2022-09-27_bill.pdf'

filePaths = [filePath_sep21, filePath_oct21, filePath_nov21, filePath_dec21, filePath_jan22, filePath_feb22, filePath_mar22, filePath_apr22, filePath_may22, filePath_jun22, filePath_jul22, filePath_aug22, filePath_sep22]
```

```python
df = wrangle.main(filePaths)
df
```

```python
df.drop(columns = 'text').to_csv('data.csv')
```
